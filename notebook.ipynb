## Part 1: Power Law PDF

The power-law probability density function is:

$$P(x) = C \cdot x^{-\alpha}, \quad x \geq x_{\min}$$

where $C = (\alpha - 1) \cdot x_{\min}^{\alpha - 1}$ is the normalization constant.

## Part 2: Maximum Likelihood Estimation

$$\hat{\alpha} = 1 + \frac{n}{\sum_{i=1}^{n} \ln(x_i / x_{\min})}$$

## Part 5: Moment Convergence (Lecture 1, Slide 9)

The $k$-th moment of a power-law distribution exists (is finite) only when $\alpha > 1 + k$:

$$\langle x^k \rangle = \int_{x_{\min}}^{\infty} x^k P(x)\,dx = \frac{(\alpha-1)\, x_{\min}^k}{\alpha - k - 1} \quad \text{for } \alpha > 1+k$$

If $\alpha \leq 1+k$, the integral diverges and the sample estimate grows with sample size $N$.
